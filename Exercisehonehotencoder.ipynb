```python
import pandas as pd
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
df1 = df.rename(columns={'Car Model':'car_model','Milage':'milage','Sell Price($)':'sell_price','Age(yrs)':'age'})
df1
```

```python
dummies = pd.get_dummies(df1.car_model)
dummies
```

```python
merged = pd.concat([df1,dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['car_model','Audi A5'],axis = 'columns')
final
```

```python
X = final.drop(['sell_price'],axis = 'columns')
X
```

```python
Y = final.sell_price
Y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X,Y)
```

```python
reg.predict([[45000,4,0,0]])
```

```python
reg.score(X,Y)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dfle = df1
dfle.car_model = le.fit_transform(dfle.car_model)
dfle
```

```python
X = dfle[['car_model','Mileage','age']].values
X
```

```python
X = ct.fit_transform(X)
X
```
